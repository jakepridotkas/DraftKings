### DK Salaries Optimizer

```python
import pandas as pd
import numpy as np
from pulp import *
```

```python
def get_pos(name):
    l = name.split('\xa0')
    return l[-1]

'Giannis Antetokounmpo PF'.split('\xa0')
```

```python
dk_data = pd.read_csv("data/numberfire_nfl.csv")
dk_data.head()
```

```python
# drop rows that have nan
dk_data = dk_data[['Player', 'Position','FP', 'Cost', 'Rank']]
dk_data = dk_data.dropna(axis=0, how='any')
dk_data.head()
```

```python
# Get the fields that are useful
#dk_data = dk_data[['Name', 'Roster Position', 'Salary', 'TeamAbbrev', 'AvgPointsPerGame']]
dk_data['Roster Position'] = dk_data['Position']#dk_data.apply(lambda x: get_pos(x['Player']),axis=1)

# change $s into int
dk_data['Cost'] = dk_data['Cost'].replace('[\$,]', '', regex=True).astype(float)
dk_data['Rank'] = dk_data['Rank'].replace('[\#,]', '', regex=True).astype(float)

dk_data.head(10)
```

```python
injured_players = ['Christian McCaffrey\xa0(RB, CAR)', 'DeAndre Hopkins\xa0(WR, ARI)']
```

```python
dk_data = dk_data[~dk_data['Player'].isin(injured_players)]
```

```python
#single game keep small list of teams

dk_data = dk_data = dk_data[dk_data.Player.str.contains("DAL\)") |
                           dk_data.Player.str.contains("DET\)") |
                           dk_data.Player.str.contains("WSH\)") |
                           dk_data.Player.str.contains("HOU\)")]
dk_data.head()
```

```python
#drop teams that aren't playing
dk_data = dk_data[~dk_data.Player.str.contains("AZ\)")]
dk_data = dk_data[~dk_data.Player.str.contains("SEA\)")]
dk_data = dk_data[~dk_data.Player.str.contains("KC\)")]
dk_data = dk_data[~dk_data.Player.str.contains("LV\)")]
dk_data = dk_data[~dk_data.Player.str.contains("LAR\)")]
dk_data = dk_data[~dk_data.Player.str.contains("TB\)")]
dk_data = dk_data[~dk_data.Player.str.contains("DAL\)")]
dk_data = dk_data[~dk_data.Player.str.contains("WAS\)")]
dk_data = dk_data[~dk_data.Player.str.contains("DET\)")]
dk_data = dk_data[~dk_data.Player.str.contains("HOU\)")]
```

```python
# Roster Positions for NBA 
nfl_roster_positions = ['QB', 'RB', 'WR', 'TE', 'D/ST']

SALARY_CAP = 50000
MAX_PLAYERS = 9


def make_vars(name, position, roster_position):
    if position in roster_position:
        return LpVariable(name + '_' + position,lowBound=0, cat='Binary')
    return 0

for position in nfl_roster_positions:
    dk_data[position] = dk_data.apply(lambda x: make_vars(x['Player'],position, x['Roster Position']),axis=1)

    #[1 if position in x else 0 for x in dk_data['Roster Position']]
```

```python
prob = LpProblem("OptimizeDK",LpMaximize)
```

```python
names = dk_data['Player']
salaries = dk_data['Cost']
projected_points = dk_data['FP']
opp_rank = dk_data['Rank']
```

```python
dk_data.head()
```

```python
# The objective function is added to 'prob' first
# OBJ FUNCTION
#prob += lpSum([dk_data[i] * projected_points + dk_data[i] * opp_rank/10 for i in nfl_roster_positions]), 'Projected Points and OppRank'
prob += lpSum([dk_data[i] * projected_points for i in nfl_roster_positions]), 'Projected Points'
```

```python
# CONSTRAINTS

# Stay under salary cap
prob += lpSum(salaries * dk_data[i] for i in nfl_roster_positions) <= SALARY_CAP

# Stay under Num Players
prob += lpSum(dk_data[i] for i in nfl_roster_positions) <= MAX_PLAYERS

# Select one of each position (flexibility for flex except QB and DST)
prob += lpSum([dk_data['QB']]) == 1
prob += lpSum([dk_data['RB']]) == 2
prob += lpSum([dk_data['WR']]) >= 3
prob += lpSum([dk_data['TE']]) >= 1
prob += lpSum([dk_data['D/ST']]) == 1

# Don't use same player twice
#for index, row in dk_data.iterrows():
#    prob += lpSum(row[i] for i in nfl_roster_positions) <= 1
```

```python
prob
```

```python
# The problem is solved using PuLP's choice of Solver
prob.solve()
```
